# The Energy Costs of LLMs
Large language models (LLMs) have become incredibly popular since their introduction in late 2023(?). LLM's are increasingly being integrated into every day tasks, but with this popularity also comes singificant rise in the energy costs associated with the AI. 

This notebook is based on an article from MIT's Technology Review, "We did the math on AI's energy footpring. Here's the story you haven't heard" by James O'Donnell and Casey Crownhart from May 20, 2025.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Text queries

```python
nparams = 8e9 # Llama 3.1 8B

njoules = 57*2
```

```python
joulesperparam = njoules/nparams
```

```python
nparams = 405e9 # Llama 3.1 405B

njoules = joulesperparam*nparams
print(njoules)
```

```python
model_params = np.array([8e9,405e9,600e9,1e12])
model_names = ["Llama 3.1 8B", "Llama 3.1 405B","DeepSeek","GPT-4"]
```

```python
joules_models = model_params*joulesperparam
```

```python
plt.scatter(model_names,joules_models)
plt.ylabel("Energy (Joules)",fontsize=18)
plt.xlabel("LLM",fontsize=18)
```
