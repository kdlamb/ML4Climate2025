# Generative Models
In this notebook we will learn about generative models. Generative models can be used to generate likely samples.

```python
import cv2
from torchvision.utils import make_grid
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
```

```python
dirpath = "/Users/karalamb/Columbia/Projects/IceVAE"
basepath = 'ICDCdataset'
shapes = ['Bud','Hoc','Loc','Plt','Cox','Hsn','Ros','Ser','Shc','Sir','Sph']
shapelabels = np.arange(len(shapes))
```

```python
cpis = []
labels = []
```

```python
for shape,label in zip(shapes,shapelabels):
    shapepath = os.path.join(dirpath,basepath,shape)
    img_list = os.listdir(shapepath)
    

    for imgs in img_list:
        img = cv2.imread(os.path.join(shapepath,imgs), cv2.IMREAD_GRAYSCALE)
        
        img = torch.tensor(1-img/img.max())
        
        cpis.append(torch.unsqueeze(img,dim=0))
        labels.append(label)
```

```python
dims = []
for i in range(0,len(cpis)):
    dims.append(np.max(cpis[i].shape))
```

```python
np.argmax(dims)
```
